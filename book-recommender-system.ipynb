## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN') # merge the ratings and books dataset on the ISBN column
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index() # group the ratings_with_name dataset by the Book-Title column and count the number of ratings
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce') # convert the Book-Rating column to numeric
ratings_with_name.head()
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index() # group the ratings_with_name dataset by the Book-Title column and calculate the mean of the Book-Rating
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True) # rename the Book-Rating column to avg_rating
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title') # merge the num_rating_df and avg_rating_df on the Book-Title column
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=300].sort_values('avg_rating',ascending=False).head(50) # filter the popular_df dataset to only include books with more than 250 ratings and sort the dataset by avg_rating in descending order
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']] # merge the popular_df and books dataset on the Book-Title column and drop duplicate rows
popular_df.head()
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 250 # group the ratings_with_name dataset by the User-ID column and count the number of ratings
padhe_likhe_users = x[x].index # filter the x dataset to only include users with more than 250 ratings
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)] # filter the ratings_with_name dataset to only include users with more than 250 ratings
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50 # group the filtered_rating dataset by the Book-Title column and count the number of ratings
```

```python
famous_books= y[y].index # filter the y dataset to only include books with more than 50 ratings
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)] # filter the filtered_rating dataset to only include books with more than 50 ratings
final_ratings  # check the final_ratings dataset 
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating') # create a pivot table from the final_ratings dataset
```

```python
pt.fillna(0,inplace=True) # fill the missing values in the pivot table with 0
```

```python
pt # check the pivot table
```

```python
from sklearn.metrics.pairwise import cosine_similarity # import the cosine_similarity function from the sklearn library
```

```python
similarity_scores = cosine_similarity(pt) # calculate the cosine similarity of the pivot table
```

```python
similarity_scores.shape # check the shape of the similarity_scores
```

```python
index = np.where(pt.index=="The Green Mile: Coffey on the Mile (Green Mile Series)")[0][0] # fetch the index of the book_name
index
```

```python
similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5] # sort the similarity_scores and fetch the top 4 similar items
similar_items[1]
```

```python
temp_df = books[books['Book-Title'] == pt.index[similar_items[1][0]]] # fetch the book details from the books dataset
temp_df
```

```python
def recommend(book_name):
    # Check if the book_name exists in the index
    if book_name not in pt.index:
        return "Book not found in the index. Please provide a valid book name."

    # Fetch the index of the book_name
    index = np.where(pt.index == book_name)[0][0]

    # Sort the similarity_scores and fetch the top 5 similar items
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        # Fetch the book details from the books dataset
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        # Append the book title, author, and image URL to the item list
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        data.append(item)

    return data
```

```python
recommend("The Kitchen God's Wife")
```
